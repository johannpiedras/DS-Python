<strong>Separating Clippers and Lakers</strong>

```python
clippers = df[df['Team'] == 'Clippers']
lakers = df[df['Team'] == 'Lakers']
```

## 4. Image
### Visualization 1
<strong>Clippers and Lakers Average Attendance</strong>

```python
fig = plt.figure()
ax = plt.gca()
#ax.set_facecolor('white')
ax.xaxis.grid()

# Clippers Graph
plt.plot(clippers['year'],clippers['road_avg'].sort_values(), ':', color='blue')
plt.plot(clippers['year'],clippers['home_avg'].sort_values(), '-',color='blue')
plt.plot(clippers['year'],clippers['overall_average'].sort_values(), '-o', color='blue')

# Lakers Graph
plt.plot(lakers['year'],lakers['road_avg'].sort_values(), ':',color='gold')
plt.plot(lakers['year'],lakers['home_avg'].sort_values(), '-', color = 'gold')
plt.plot(lakers['year'],lakers['overall_average'].sort_values(), '-o', color = 'gold')

# Aesthetics
plt.xlabel('Year')
plt.ylabel('Attendance')
plt.title('Clippers and Lakers Average Attendance')
plt.legend(['Clippers Road Avg', 'Clippers Home Avg','Clippers Overall Avg', 'Lakers Road Avg', 'Lakers Home Avg','Lakers Overall Avg']);

# remove the frame of the chart
for spine in plt.gca().spines.values():
    spine.set_visible(False)
plt.tight_layout()

plt.show()
```

<strong>EXTRA: Clippers and Lakers Total Home Attendance</strong>

```python

fig = plt.figure()
ax = plt.gca()


plt.plot(clippers['year'],clippers['home_total'].sort_values(), '-o', color = 'blue')
plt.plot(lakers['year'],lakers['home_total'].sort_values(), '-o', color = 'gold')

plt.xlabel('Year')
plt.ylabel('Total Home Attendance')
plt.title('Clippers and Lakers Total Home Attendance')
plt.legend(['Clippers', 'Lakers']);

# remove the frame of the chart
for spine in plt.gca().spines.values():
    spine.set_visible(False)
    
ax.yaxis.grid()
plt.tight_layout()
    
```

## 5. Discussion
